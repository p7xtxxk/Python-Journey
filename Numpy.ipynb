```python
import numpy as np
print(np.__version__)
```

```python
# working with numPy arrays
a = np.array([1, 2, 3, 4, 5, 6, 7])
print(a)
print(type(a))
a *= 2
print(a)
print(a.ndim)


# 2 dimensional array
b = np.array([[1, 2, 3, 4],
              [2, 3, 4, 5],
              [3, 4, 5, 6]])
print(b)
print(b.ndim)

# 3d array
c = np.array([[[1, 2, 3], [2, 3, 4], [3, 4, 5]], [[1, 2, 3], [
             2, 3, 4], [3, 4, 5]], [[4, 5, 6], [5, 6, 7], [6, 7, 8]]])
print(c.ndim)
print(c.shape)

# multi dimensional indexing
# 1d
print(a[0])
print(a[2])
#2d
print(b[0,1])
print(b[2,3])
print(b[1])
#3d
print(c[0,0,0]) #specific element
print(c[2,1]) # specific row
print(c[1]) # specific layer
```

```python
# slicing in NumPy array
d = np.array([[1, 2, 3, 4],
              [5, 6, 7, 8],
              [2, 4, 6, 8],
              [1, 3, 5, 7]])
# print(d)
# print(d.ndim)
# print(d.shape)
# print all elements
print(d[:])
# row selection
print(d[::2])
print(d[::-2]) # return every 2nd row in reverse
print(d[::-1]) # return the whole in reverse
print(d[:3])
```

```python
# # column selection
# print(d[::, ::-1])

# row and column selection
print(d[::-2, ::-2])
```

```python
#arithmetic operations
print(d+2)
print(d-2)
print(d*3)
print(d/2)
print(d%2)
print(d**2)
print(d//2)
```

```python
# vectorized math functions
e = np.array([1.2, 4.55, 7.8822, 7.992])
print(np.square(e))
print(np.sqrt(e))
print(np.round(e))
print(np.floor(e))
print(np.ceil(e))
print(np.pi)
```

```python
#  element wise arithmetic
a1 = np.array([1, 2, 3, 4, 5])
a2 = np.array([5, 6, 7, 8, 9])
print(a1+a2)
print(a1-a2)
print(a1*a2)
print(a1/a2)
print(a1 % a2)
print(a1**a2)
print(a1//a2)
```

```python
# comparison operators
s = np.array([22,442,332,23,23422,2342,34,2,2415567,4,64,4,22])
print(s>100)
print(s==100)
print(s<100)
print(s<=100)
print(s>=100)

# we can assign value to specific places too
s[s<1000] = 1234
print(s)
```
